# Regression

```python
final = pd.read_csv("C:/Users/DELL/Downloads/Datasets/finalagriculture.csv")
final.head()
```

```python
final.columns
```

```python
final.isnull().sum()
```

```python
final['Production']=final['Production'].fillna(final['Production'].mean())
final['pressure']=final['pressure'].fillna(final['pressure'].mean())
final['visibility']=final['visibility'].fillna(final['visibility'].mean())
```

```python
final.isnull().sum().sum()
```

```python
train=final[['Area',
        'pressure', 'temperatureMax',
       'visibility', 'windBearing', 'windSpeed']]
test=final[['Yield']]
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from scipy import stats
import statsmodels.api as sm
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=30,random_state=1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,accuracy_score,r2_score
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
X2 = sm.add_constant(train)
est = sm.OLS(test, X2)
est2 = est.fit()
print(est2.summary())
```

```python
#print(r.summary())
```

```python
predict=regressor.predict(x_test)
print(predict)
```

```python
print("mean squared error is:",mean_squared_error(y_test,predict))
print("r2 is :",r2_score(y_test,predict))
#print("Accuracy:",accuracy_score(y_test,predict))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import f1_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=30,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
regr=DecisionTreeRegressor(random_state=1,criterion='mse',min_samples_split=2)
regr.fit(x_train,y_train)
```

```python
p=regr.predict(x_test)
```

```python
print("mean_squared_error:",mean_squared_error(y_test,p))
print("r2_square:",r2_score(y_test,p))
#print("f1_score:",f1_score(y_test,p))
```
