```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
eda['Production']=eda['Production'].fillna(eda['Production'].mean())
eda['pressure']=eda['pressure'].fillna(eda['pressure'].mean())
eda['visibility']=eda['visibility'].fillna(eda['visibility'].mean())
```

```python
eda['Season'].unique()
```

```python
plt.figure(figsize = (6,6))
plot_shipmode = eda['Season'].value_counts()
colors = ('Gold' , 'Salmon' , 'Cyan' , 'LightGreen')
label = eda['Season'].unique()

plt.pie(plot_shipmode, 
       autopct = '%1.1f%%',
       explode = (0.08,0.08,0.08,0.08),
       shadow = True,
       colors = colors,
       labels = label);
```

```python
sns.catplot(data = eda , x = 'District_Name' , aspect = 3 , kind = 'count')
plt.xticks(rotation = 90);
```
