```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
label=LabelEncoder()
iris['Species']=label.fit_transform(iris['Species'])
```

```python
iris['Species'].tail()
```

```python
final=iris.drop(['Id'],axis=1)
```

```python
train=final[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
test=final[['Species']]
```

```python
# train=final.drop(columns=['Species'],axis=1)
# test=final['Species']
```

```python
test.head()
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=30,random_state=0)
x_train.shape,y_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
classi=GaussianNB()
classi.fit(x_train,y_train)
```

```python
predict=classi.predict(x_test)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy is:",accuracy_score(y_test,predict))
print("F1 Score is:",classification_report(y_test,predict))
print("confusion matrix:",confusion_matrix(y_test,predict))
```
